```python
#Read in the data using pandas
#using a variable called as df to read the datapath file, which has excel sheet table with the dataset
#using the pandas library retrive the dataset to the system , through the excel sheet root path.
df = pd.read_csv('/content/drive/MyDrive/Machine_Learning_CourseWork/5DATA002W.2 Coursework Dataset(25012025v6.0) (2).csv')
```

```python






#%% Initial Data Exploration (Reused from Seminar 3)
# Display first 5 rows
print("Initial Data Preview:")
print(df.head())







#%% Handle Missing Values (Reused from Seminar 4)
# Check for missing values
print("\nMissing Values Before Handling:")
print(df.isnull().sum())



# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)



# Drop rows with missing Tumor_Size
df = df.dropna(subset=['Tumor_Size'])

# Verify missing values handled
print("\nMissing Values After Handling:")
print(df.isnull().sum())



#%% Handle Invalid Values (Reused from Seminar 5)
# Fix negative age
df['Age'] = df['Age'].abs()



# Fix negative tumor size
df['Tumor_Size'] = df['Tumor_Size'].abs()

# Remove unrealistic ages (>100)
df = df[df['Age'] <= 100]



#%% Feature Engineering (Reused from Seminar 6)
# Create Age Groups
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[0, 40, 60, 100],
                         labels=['Young', 'Middle', 'Senior'])


#  Reused from Code Reuse Session 1 (Week 5) - Page 2
#  Reused from Seminar 6 - Age Group Creation
# âž¤Custom logic approved by [Tutor Name] in Peer Review 1


# One-hot encode categorical features
cat_cols = ['Sex', 'T_Stage', 'N_Stage', '6th_Stage', 'Differentiated', 'A_Stage']
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)


#%% Create Classification Dataset (Reused from Code Reuse Session 1)
# Classification dataset for mortality prediction
classification_df = df_encoded.copy()


# Drop irrelevant columns
classification_df = classification_df.drop(['Survival_Months', 'Patient_ID'], axis=1)


# Convert target to binary
classification_df['Mortality_Status'] = classification_df['Mortality_Status'].map({'Alive':0, 'Dead':1})



#%% Create Regression Dataset (Reused from Code Reuse Session 1)
# Regression dataset for survival months prediction
regression_df = df_encoded.copy()


# Drop irrelevant columns
regression_df = regression_df.drop(['Mortality_Status', 'Patient_ID'], axis=1)




#%% Save Processed Datasets (Reused from Seminar 8)
# Save classification datas
classification_df.to_csv(os.path.join(save_path, 'classification_dataset3.csv'), index=False)



# Save regression dataset
regression_df.to_csv('regression_dataset.csv', index=False)


print("\nPreprocessing completed. Datasets saved.")
```
